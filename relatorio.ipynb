O desvio padrão das avaliações é 0.63. Isso indica que as avaliações dos episódios variam em torno da média de forma relativamente consistente. Ou seja, embora a maioria dos episódios tenha uma avaliação próxima da média, ainda há uma certa variação nas opiniões dos espectadores.

## Relação entre variáveis

Analise a relação entre as variáveis referentes a cada episódio. Explore visualmente a relação entre elas e calcula estatísticas de correlação.

### Existe alguma relação entre a quantidade de votos recebida e a avaliação média dos episódios?

```python
# INSIRA SEU CÓDIGO AQUI

correlacao_quantvotos_avmedia = episodios['num_votos'].corr(episodios['avaliacao_media'])

print(f"Correlação entre quantidade de votos e avaliação média: {correlacao_quantvotos_avmedia:.2f}")
```

```python
from scipy.stats import kendalltau

pearson_corr = episodios['num_votos'].corr(episodios['avaliacao_media'])
spearman_corr = episodios['num_votos'].corr(episodios['avaliacao_media'], method='spearman')
kendall_corr, _ = kendalltau(episodios['num_votos'], episodios['avaliacao_media'])

plt.figure(figsize=(10, 6))
plt.scatter(episodios['num_votos'], episodios['avaliacao_media'], alpha=0.5)
plt.title('Relação entre quantidade de votos e avaliação média dos episódios')
plt.xlabel('Quantidade de votos')
plt.ylabel('Avaliação média dos episódios')
plt.grid(True)
plt.show()

print(f"Correlação de Pearson: {pearson_corr}")
print(f"Correlação de Spearman: {spearman_corr}")
print(f"Correlação de Kendall: {kendall_corr}")
```

Podemos observar que existe uma relação moderadamente positiva entre o número de votos e a avaliação média dos episódios. Isso significa que, em geral, episódios com um maior número de votos tendem a receber avaliações médias mais altas.

Os valores de correlação reforçam essa associação:

Correlação de Pearson: 0.316
Correlação de Spearman: 0.469
Correlação de Kendall: 0.332
Uma correlação de 0.5 no coeficiente de Pearson indica uma correlação positiva moderada entre as duas variáveis. Isso sugere que, em geral, quando o número de votos aumenta, a avaliação média também tende a aumentar, mas essa relação não é perfeita.

Essa tendência é consistente nas diferentes métricas de correlação utilizadas: Pearson, Spearman e Kendall. Você pode visualizar essa relação no gráfico abaixo.

### Os episódios no fim das temporadas são melhor avaliados do que os do início?

```python
# INSIRA SEU CÓDIGO AQUI

temporada_episodio_avaliacao = episodios.groupby(['temporada', 'num_episodio'])['avaliacao_media'].mean()

for temporada, grupo in temporada_episodio_avaliacao.groupby('temporada'):
    primeiros = grupo.head(3).values[0]
    ultimos = grupo.tail(3).values[0]
    
    if ultimos > primeiros:
        print(f"Na temporada {temporada}, os episódios no fim da temporada ({primeiros:.2f}) são melhor avaliados do que os do início ({primeiros:.2f}).")
    elif ultimos < primeiros:
        print(f"Na temporada {temporada}, os episódios no início da temporada ({ultimos:.2f}) são melhor avaliados do que os do fim ({ultimos:.2f}).")
    else:
        print(f"Na temporada {temporada}, os episódios no início e no fim da temporada têm a mesma média de avaliação ({primeiros:.2f}).")

```

```python
avaliacao_inicio_temporada = episodios[episodios['num_episodio'] <= 5]['avaliacao_media'].mean()
avaliacao_fim_temporada = episodios[episodios['num_episodio'] > episodios['num_episodio'].max() - 5]['avaliacao_media'].mean()

plt.figure(figsize=(8, 6))
plt.bar(['Início da Temporada', 'Fim da Temporada'], [avaliacao_inicio_temporada, avaliacao_fim_temporada], color=['blue', 'orange'])
plt.title('Média de Avaliação dos Episódios no Início e Fim das Temporadas')
plt.ylabel('Avaliação Média')
plt.ylim(0, 10)
plt.grid(axis='y')
plt.show()
```

As avaliações dos episódios tendem a ser maiores no final das temporadas em comparação com o início. Essa tendência sugere que os episódios finais das temporadas podem ser percebidos como mais impactantes, emocionantes ou satisfatórios pelo público em comparação com os episódios iniciais. 

```python
temporada_episodio_avaliacao = episodios.groupby(['temporada', 'num_episodio'])['avaliacao_media'].mean()

primeiros_episodios = temporada_episodio_avaliacao.groupby('temporada').head(3)
ultimos_episodios = temporada_episodio_avaliacao.groupby('temporada').tail(3)

plt.figure(figsize=(10, 6))
temporadas = primeiros_episodios.index.get_level_values('temporada').unique()

for temporada in temporadas:
    plt.plot(primeiros_episodios.loc[temporada].index, primeiros_episodios.loc[temporada].values, marker='o', label=f'Temporada {temporada} - Primeiros', linestyle='-')

for temporada in temporadas:
    plt.plot(ultimos_episodios.loc[temporada].index, ultimos_episodios.loc[temporada].values, marker='o', label=f'Temporada {temporada} - Últimos', linestyle='--')

plt.title('Média de Avaliação dos Primeiros e Últimos 3 Episódios por Temporada')
plt.xlabel('Episódio')
plt.ylabel('Média de Avaliação')
plt.legend()
plt.grid(True)

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```

### Existem episódios outliers em termos de popularidade?

```python
# INSIRA SEU CÓDIGO AQUI

media_votos = episodios['num_votos'].mean()
desvio_padrao_votos = episodios['num_votos'].std()

limite_inferior = media_votos - 2 * desvio_padrao_votos
limite_superior = media_votos + 2 * desvio_padrao_votos

outliers_popularidade = episodios[(episodios['num_votos'] < limite_inferior) | (episodios['num_votos'] > limite_superior)]

if not outliers_popularidade.empty:
    print("Episódios outliers identificados em termos de popularidade:")
    print(outliers_popularidade)
else:
    print("Não foram encontrados episódios outliers em termos de popularidade.")
```

```python
outliers_popularidade = episodios[(episodios['num_votos'] < limite_inferior) | (episodios['num_votos'] > limite_superior)]

plt.figure(figsize=(10, 6))
plt.bar(outliers_popularidade['nome_episodio'], outliers_popularidade['num_votos'], color='red')
plt.title('Episódios Outliers em Termos de Popularidade')
plt.xlabel('Episódio')
plt.ylabel('Quantidade de Votos')
plt.tight_layout()
plt.show()
```

Descobrimos dois episódios da série "Stranger Things" que se comportam como verdadeiros outliers em termos de popularidade, ambos da 4ª temporada:

Capítulo Quatro: Caro Billy: este episódio conquistou o público com 37.149 votos, um número impressionante!
Capítulo Sete: O Massacre no Laboratório Hawkins: com 40.864 votos, este episódio se tornou um verdadeiro campeão de popularidade.
Comparando com os demais episódios da série, a quantidade de votos recebidos por esses dois episódios é consideravelmente maior, indicando que algo especial os torna tão atraentes para os fãs.

### Existem episódios outliers em termos de avaliação recebida?

```python
# INSIRA SEU CÓDIGO AQUI

primeiro_quartil = episodios['avaliacao_media'].quantile(0.25)
terceiro_quartil = episodios['avaliacao_media'].quantile(0.75)

IQR = terceiro_quartil - primeiro_quartil

limite_inferior = primeiro_quartil - 1.5 * IQR
limite_superior = terceiro_quartil + 1.5 * IQR

outliers = episodios[(episodios['avaliacao_media'] < limite_inferior) | (episodios['avaliacao_media'] > limite_superior)]

if not outliers.empty:
    print("Episódios outliers identificados:")
    print(outliers)
else:
    print("Não foram encontrados episódios outliers.")
```

```python
outliers = episodios[(episodios['avaliacao_media'] < limite_inferior) | (episodios['avaliacao_media'] > limite_superior)]
nao_outliers = episodios[~((episodios['avaliacao_media'] < limite_inferior) | (episodios['avaliacao_media'] > limite_superior))]

plt.figure(figsize=(10, 6))
plt.scatter(nao_outliers['num_episodio'], nao_outliers['avaliacao_media'], color='blue', label='Episódios não outliers')
plt.scatter(outliers['num_episodio'], outliers['avaliacao_media'], color='red', label='Episódios outliers')
plt.title('Distribuição das Avaliações dos Episódios')
plt.xlabel('Número do Episódio')
plt.ylabel('Avaliação Média')
plt.legend()
plt.grid(True)
plt.show()
```

Após a análise dos dados, identificamos um episódio considerado um outlier em termos de avaliação recebida. Este episódio, intitulado "Chapter Seven: The Lost Sister" da série "Stranger Things", da temporada 2, recebeu uma avaliação média de 6.1, destacando-se significativamente em relação à distribuição geral das avaliações dos outros episódios.
